```python
import pandas as pd
inputfile = "https://dr3vr6j2erh62.cloudfront.net/mediastore/dsm020/datasets/IRIS.csv"

df = pd.read_csv(inputfile)

df.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
species = list(set(df["species"]))
counts = [df.loc[df["species"]==name].size for name in species]
explode=(0, 0.1, 0.2)

plt.figure()

plt.pie(counts,
        explode = explode,
       labels = species,
       autopct = "%1.2f")

plt.show()
```
